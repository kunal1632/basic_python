```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
x=np.arange(0,5* np.pi, 0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
```

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color ='g')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
